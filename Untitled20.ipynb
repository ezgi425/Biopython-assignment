```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
import os

fasta_files = [f for f in os.listdir() if f.endswith(".fasta")]

for fasta in fasta_files:
    records = list(SeqIO.parse(fasta, "fasta"))

    for record in records:
        record.annotations["molecule_type"] = "protein"
        record.annotations["organism"] = "Homo sapiens"
        record.annotations["data_file_division"] = "PRI"
        record.annotations["date"] = "01-JAN-2000"

    output_file = fasta.replace(".fasta", ".gb")
    SeqIO.write(records, output_file, "genbank")

    print(f"Oluşturuldu: {output_file}")
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_protein
import os

fasta_files = [f for f in os.listdir() if f.endswith(".fasta")]

for fasta in fasta_files:
    records = []

    for record in SeqIO.parse(fasta, "fasta"):
        print(record.id)

        new_seq = Seq(str(record.seq), generic_protein)

        new_record = SeqRecord(
            new_seq,
            id=record.id,
            name=record.name,
            description=record.description
        )

        new_record.annotations["molecule_type"] = "protein"
        records.append(new_record)

    output_file = fasta.replace(".fasta", ".gb")
    SeqIO.write(records, output_file, "genbank")
    
    print(f"Oluşturuldu: {output_file}")
```
